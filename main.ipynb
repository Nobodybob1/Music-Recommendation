```python
from sklearn.preprocessing import StandardScaler

features = ['valence', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'loudness', 'tempo']
X = train_data[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=6, metric='cosine')
model.fit(X_scaled)
```

```python
from difflib import get_close_matches

def find_similar_tracks(track_name, train_data):
    names = get_close_matches(track_name, train_data['name'].tolist(), n=3, cutoff=0.6)
    if not names:
        return "Track not found."
    return names
```

```python
find_similar_tracks('toxcty', train_data)
```

```python
def recommend_songs(name, train_data, model, X_scaled):
    if name not in train_data['name'].values:
        names = find_similar_tracks(name, train_data)
        for n in names:
            response = input(f"Did you mean: {n}? (y/n) ").strip().lower()
            if response == 'y':
                name = n
                break
            elif response == 'n':
                if n == names[-1]:
                    print("No similar tracks found. Please try another name.")
                    return None
                continue
            else:
                print("Invalid response.")
                return None
    
    song_index = train_data[train_data['name'] == name].index[0]
    distances, indices = model.kneighbors(X_scaled[song_index].reshape(1, -1))

    print(f"Recommendations for '{name}':")
    for i in range(1, len(distances[0])):
        recommended_song_index = indices[0][i]
        recommended_song = train_data.iloc[recommended_song_index]
        print(f"{i}. {recommended_song['name']} by {recommended_song['artists']} (Distance: {distances[0][i]:.4f})")
    return indices[0][1:]  # Return indices of recommended songs
```

```python
recommend_songs('toxity', train_data, model, X_scaled)
```
